```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("dog.jpg")
```

```python
img.dtype
```

```python
plt.imshow(img)
plt.show()
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
img_rgb.shape
```

```python
x, y, h = 171, 294, 3


img_rgb_copy = [[list(pixel) for pixel in row] for row in img_rgb]

for i in range(x):
    for j in range(y):
        for k in range(h):
            img_rgb[i][j][k] = min(img_rgb_copy[i][j][k] - 50, 255)

plt.imshow(img_rgb)
plt.axis('off')  
plt.show()
```

#   SHIFTING THE IMAGE

```python
img_rgb_copy = cv2.resize(img_rgb, (500, 500))

for i in range(min(x, img_rgb_copy.shape[0])):
    for j in range(min(y, img_rgb_copy.shape[1])):
        for k in range(h):
            img_rgb_copy[i][j][k] = img_rgb_copy[i][min(j + 50, img_rgb_copy.shape[1] - 1)][k]

plt.imshow(img_rgb_copy)
plt.show()
```
