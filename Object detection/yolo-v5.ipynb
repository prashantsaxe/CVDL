```python
import shutil

# Define paths
data_yaml_path = "/kaggle/input/traffic-detection-project/data.yaml"
new_data_yaml_path = "/kaggle/working/data.yaml"

# Copy the file to a writable location
shutil.copy(data_yaml_path, new_data_yaml_path)
```

```python
import yaml

data_yaml_path = "/kaggle/working/data.yaml"

with open(data_yaml_path, "r") as f:
    data = yaml.safe_load(f)

# Update paths
data['train'] = "/kaggle/input/traffic-detection-project/train/images"
data['val'] = "/kaggle/input/traffic-detection-project/valid/images"
data['test'] = "/kaggle/input/traffic-detection-project/test/images"

# Save changes
with open(data_yaml_path, "w") as f:
    yaml.dump(data, f)

print("data.yaml updated successfully!")
```
